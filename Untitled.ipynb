```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
def calculate_value_percentages_and_fill_nans(df, column_name):
    col = df[column_name]
    non_zero = col[col != 0]
    if len(non_zero) == 0:
        return {val: 0 for val in col.unique()}
    
    counts = non_zero.value_counts()
    total = counts.sum()
    percentages = {val: (count / total) * 100 for val, count in counts.items()}
    
    missing = df[column_name].isna()
    if missing.any():
        choices = list(percentages.keys())
        probs = [percentages[c] / 100 for c in choices]
        df.loc[missing, column_name] = np.random.choice(choices, size=missing.sum(), p=probs)
    
    return percentages

calculate_value_percentages_and_fill_nans(df_train, 'Time_spent_Alone')
calculate_value_percentages_and_fill_nans(df_test, 'Time_spent_Alone')
calculate_value_percentages_and_fill_nans(df_train, 'Stage_fear')
calculate_value_percentages_and_fill_nans(df_test, 'Stage_fear')
calculate_value_percentages_and_fill_nans(df_train, 'Social_event_attendance')
calculate_value_percentages_and_fill_nans(df_test, 'Social_event_attendance')
calculate_value_percentages_and_fill_nans(df_train, 'Going_outside')
calculate_value_percentages_and_fill_nans(df_test, 'Going_outside')
calculate_value_percentages_and_fill_nans(df_train, 'Drained_after_socializing')
calculate_value_percentages_and_fill_nans(df_test, 'Drained_after_socializing')
calculate_value_percentages_and_fill_nans(df_train, 'Friends_circle_size')
calculate_value_percentages_and_fill_nans(df_test, 'Friends_circle_size')
calculate_value_percentages_and_fill_nans(df_train, 'Post_frequency')
calculate_value_percentages_and_fill_nans(df_test, 'Post_frequency')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in df_test.columns:
    try:
        if df_train[i].dtype == object:
            label = LabelEncoder()
            df_train[i] = label.fit_transform(df_train[i])
            df_test[i] = label.transform(df_test[i])
    except:
        print(i)
```

```python
sns.boxplot(x=df_test['Post_frequency'])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_estimators':[20, 30, 50, 70],
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_depth':[6, 8, 10, 11],
    'max_features':['sqrt', 'log2', None]
}
```

```python
final_model = GridSearchCV(model, param_grid=params, cv=3, verbose=2, return_train_score=True)
```

```python
X_train = df_train.drop('Personality', axis=1)
y_train = df_train['Personality']
X_test = df_test.copy()
```

```python
final_model.fit(X_train, y_train)
```

```python
final_model.best_params_
```

```python
final_model.best_score_
```

```python
y_pred = final_model.predict(X_test)
```
